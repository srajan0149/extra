### 1 ###

Let $A_{n\times k}$ and $B_{k\times m}$ be the two matrices we have to multiply.

When we multiply a single row of $A$ with a single column of $B$, we will need $k$ multiplications, so for $n$ rows of $A$ and $m$ columns of $B$, we will need $k\times m\times n$ mulitplications.

If we have to add $l$ number of terms we need $l-1$ additions, so for $n$ rows and $m$ columns with $k$ terms as each element, we will have to do $(k-1)\times m\times n$ additions.

```python
import time
import random
import numpy as np

(n,k,m) = (random.randint(1,100),random.randint(1,100),random.randint(1,100))
print(f"n: {n}, k: {k}, m: {m}")
A = [[random.random() for i in range(k)] for j in range(n)]
B = [[random.random() for i in range(m)] for j in range(k)]

t0 = time.time()
C = [[0 for i in range(m)] for j in range(n)]
for i in range(n):
    for j in range(m):
        for l in range(k):
            C[i][j] += A[i][l]*B[l][j]
t1 = time.time()
print("Time taken by using List of Lists: ",t1-t0)

A_np = np.array(A)
B_np = np.array(B)

t2 = time.time()
C = A_np@B_np
t3 = time.time()
print("Time taken by using Numpy Array: ",t3-t2)
```

The numpy matrix multiplication is faster than the usual matrix multiplication as numpy arrays are of homogenous type and are stored in a contiguous memory and so does not involve type checking for each element and can loop faster through elements.

```python
import matplotlib.pyplot as plt
import numpy as np

X = np.arange(0.5,100.0,0.5)

fig, ax = plt.subplots(figsize=(10,20),ncols=2,nrows=4)

ax[0][0].plot(X,X)
ax[0][0].set_title("$y=x$")

ax[0][1].plot(X,X*X)
ax[0][1].set_title("$y=x^2$")

ax[1][0].plot(X,X*X*X/100)
ax[1][0].set_title("$y={x^3}/100$")

ax[1][1].plot(X,np.sin(X))
ax[1][1].set_title("$y=sin(x)$")

ax[2][0].plot(X,X*X)
ax[2][0].set_title("$y=sin(x)/x$")

ax[2][1].plot(X,np.log(X))
ax[2][1].set_title("$y=log(x)$")

ax[3][0].plot(X,np.exp(X))
ax[3][0].set_title("$y=e^x$")

plt.legend()
plt.show()
```
